```python
from google.colab import userdata
key = userdata.get('GEMINI_API_KEY')
```

```python
import base64
from google import genai

client = genai.Client(api_key = key)
```

```python
with open("/content/sample_data/sample.jpg", "rb") as f:
  image_byte = f.read()

# for security the encoding and decoding is done
image_b64 = base64.b64encode(image_byte).decode("utf-8")
```

```python
response = client.interactions.create(
    model = "gemini-3.5-flash",
    input = [
        {
            "type": "image",
            "data": image_b64,
            "mime_type": "image/jpeg"
        },
        {
            "type": "text",
            "text": "Explain about this image"
        }
    ]
)
```

```python
response.output_text
```

```python
response2 = client.interactions.create(
    model = "gemini-3.5-flash",
    input = "What would you add in this image",
    previous_interaction_id = response.id
)
```
